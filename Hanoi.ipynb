# 绘制解决方案图像并生成视频
p.s.注意运行前建立"../hanoi/imgs/"文件夹并保证文件夹下为空

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import os


def draw(left, center, right):
    
    fig, axs = plt.subplots(1, 3, figsize=(15, 2.5))
    
    base_x = 0.5
    block_height = 0.12

    for i in range(7):
        block_y = block_height * i
        for j, ls in enumerate([left, center, right]):
            block_width = ls[i] * 0.1
            block_x = base_x - block_width / 2
            block = patches.Rectangle((block_x, block_y), block_width, block_height, edgecolor='black', linewidth=2, facecolor='tan')
            axs[j].add_patch(block)
    
    [ax.axis('off') for ax in axs]
    count = 0
    while os.path.exists("../hanoi/imgs/{}.jpg".format(count)):
        count += 1
    plt.savefig("../hanoi/imgs/{}.jpg".format(count))
    plt.show()
    plt.clf()
    

def action(state, start, end):
    location_map = {"left": 0, "center": 1, "right": 2}
    tmp = state[location_map[start]][-1]
    state[location_map[start]] = state[location_map[start]][:-1]
    state[location_map[end]].append(tmp)
    padded_state = pad(state, 7)
    draw(*padded_state)
    return state
    
    
def pad(state, length):
    padded_state = [example_list + [0] * (length - len(example_list)) for example_list in state]
    return padded_state


def hanoi(n, start, tmp, end, state):
    if n == 1:
        print(start, "-->", end)
        state = action(state, start, end)
    else:
        hanoi(n-1, start, end, tmp, state)
        print(start, "-->", end)
        state = action(state, start, end)
        hanoi(n-1, tmp, start, end, state)

    
left = [7,6,5,4,3,2,1]
center = []
right = []
state = [left, center, right]
draw(*pad(state, 7))
hanoi(7, "left", "right", "center", state)
```

```python
import os
import imageio
# from PIL import Image

# set file name and path for the video
filename = 'output.mp4'
filepath = "../hanoi/"

# read all pictures
input_path = "../hanoi/imgs/"
img_paths = os.listdir(input_path)
img_paths.sort(key=lambda x:int(x[:-4]))
mp4_images = []
for path in img_paths:
    mp4_images.append(imageio.imread(input_path + path))

# turn .jpg to mp4
fps = 5  # 10 frame per second
with imageio.get_writer(filepath + filename, fps=fps) as video:
    for image in mp4_images:
        # frame = image.convert('RGB')
        video.append_data(image)
```
